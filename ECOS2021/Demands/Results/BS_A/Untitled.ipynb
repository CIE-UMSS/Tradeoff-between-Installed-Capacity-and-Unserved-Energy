```python
import pandas as pd
import numpy as np

Jan=pd.read_csv('Jan_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Feb=pd.read_csv('Feb_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Mar=pd.read_csv('Mar_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Apr=pd.read_csv('Apr_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
May=pd.read_csv('May_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Jun=pd.read_csv('Jun_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Jul=pd.read_csv('Jul_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Aug=pd.read_csv('Aug_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Sep=pd.read_csv('Sep_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Oct=pd.read_csv('Oct_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Nov=pd.read_csv('Nov_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
Dec=pd.read_csv('Dec_BS_A.csv', sep=',', decimal=',', encoding='latin1', error_bad_lines=False)
p=pd.concat([Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec],ignore_index=True)

del p['Unnamed: 0']
p.to_csv('minutely_BS_A.csv')
```
